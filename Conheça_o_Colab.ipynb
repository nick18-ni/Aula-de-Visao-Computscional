# Classifica√ß√£o com IA

```
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing import image
```

```
# CNN com Dropout para Redu√ß√£o de Overfitting
# Overfitting acontece quando um modelo aprende bem demais os dados de treino, a ponto de memorizar os exemplos, mas n√£o consegue generalizar para novos dados (como os de valida√ß√£o ou teste).
# Dropout √© uma t√©cnica de regulariza√ß√£o usada em redes neurais para reduzir o overfitting.Durante o treinamento, o Dropout desliga aleatoriamente uma porcentagem dos neur√¥nios em uma camada.
# Este script realiza o treinamento de uma CNN no dataset CIFAR-10
# Inclui camadas de Dropout para combater overfitting e plota os resultados.

import matplotlib.pyplot as plt                   # Criar gr√°ficos e visualiza√ß√µes.
import tensorflow as tf                           # Criar e treinar redes neurais. Usamos para montar a CNN com Dropout, carregar o dataset CIFAR-10 e treinar o modelo.
from tensorflow.keras import layers, models       # √â a interface do TensorFlow para criar modelos de forma mais simples.

# Carregando o dataset CIFAR-10, que cont√©m 60 mil imagens (32x32x3) de 10 classes diferentes
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

# Normalizando os dados para facilitar o aprendizado da rede
# Convertendo de inteiros (0-255) para floats (0.0 at√© 1.0)
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Definindo os nomes das classes (apenas para refer√™ncia/visualiza√ß√£o)
classes = ['avi√£o', 'carro', 'p√°ssaro', 'gato', 'cervo', 'cachorro', 'sapo', 'cavalo', 'navio', 'caminh√£o']

# Constru√ß√£o da arquitetura da rede neural convolucional com Dropout
# O Dropout ajuda a prevenir o overfitting desligando aleatoriamente neur√¥nios durante o treinamento
modelo_dropout = models.Sequential([

    # Conv2D: Aplica filtros (kernels) para detectar padr√µes como bordas e formas.
    # 32: N√∫mero de filtros (vai gerar 32 imagens de sa√≠da por entrada).
    # (3, 3): Tamanho do filtro (3x3 pixels).
    #activation='relu'`: Fun√ß√£o de ativa√ß√£o que mant√©m valores positivos e zera os negativos.
    #input_shape=(32, 32, 3): Formato da imagem de entrada (32x32 pixels, 3 canais de cor ‚Äì RGB).

    # Primeira camada convolucional + pooling
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),  # Reduz a imagem pela metade (2x2). Mantem os pontos importantes

    # Segunda camada convolucional + pooling
    layers.Conv2D(64, (3, 3), activation='relu'), #aprende padr√µes mais complexos.
    layers.MaxPooling2D((2, 2)),

    # Terceira camada convolucional (sem pooling aqui).
    layers.Conv2D(64, (3, 3), activation='relu'),


    # Flatten para transformar a imagem em vetor (necess√°rio para Dense que usa vetor como entrada)
    layers.Flatten(),

    # Camada densa totalmente conectada com 64 neur√¥nios, s√≥ vai ativar os neur√¥nios se os valores forem positivos. N√∫meros negativos geralmente n√£o carregam informa√ß√µes √∫teis
    layers.Dense(64, activation='relu'),

    # Dropout com taxa de 50% (desliga metade dos neur√¥nios aleatoriamente), evita que o modelo dependa demais de alguns neur√¥nios ‚Üí combate o overfitting.
    layers.Dropout(0.5),

    # Camada de sa√≠da com 10 neur√¥nios (uma para cada classe), com softmax para classifica√ß√£o. Transforma os valores de sa√≠da em probabilidades entre 0 e 1.
    layers.Dense(10, activation='softmax')
    #  O modelo vai prever qual classe tem maior probabilidade.

])

# Essa fun√ß√£o compara a sa√≠da softmax com o n√∫mero da classe correta (0 a 9) e calcula o erro.
modelo_dropout.compile(optimizer='adam', # decide como ajustar os pesos
                       loss='sparse_categorical_crossentropy',
                       metrics=['accuracy'])

# Treinando o modelo por 10 √©pocas, validando a cada √©poca nos dados de teste
hist_dropout = modelo_dropout.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))

# Plotando o gr√°fico de acur√°cia (treino e valida√ß√£o) para analisar o overfitting
plt.plot(hist_dropout.history['accuracy'], label='Treino com Dropout')
plt.plot(hist_dropout.history['val_accuracy'], label='Valida√ß√£o com Dropout')
plt.xlabel('√âpoca')
plt.ylabel('Acur√°cia')
plt.ylim([0.5, 1]) # entre 50% a 100% de acerto
plt.legend(loc='lower right')
plt.title('Acur√°cia de Treino vs Valida√ß√£o (com Dropout)')


# acur√°cia: mede quantas previs√µes o modelo acertou
# 'accuracy': acur√°cia no conjunto de treino
# 'val_accuracy': acur√°cia no conjunto de valida√ß√£o
# 'loss': erro no treino
# 'val_loss': erro na valida√ß√£o

# Avaliando o modelo final nos dados de teste
test_loss, test_acc = modelo_dropout.evaluate(x_test, y_test, verbose=2)
print(f"Acur√°cia final no teste: {test_acc:.4f}")
```
